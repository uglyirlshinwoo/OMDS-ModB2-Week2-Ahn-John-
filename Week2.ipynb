```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

df = pd.read_csv('brca_metabric_clinical_data.tsv',sep='\t')

# Step 1: Summarize the Dataset

print("Dataset Info:")
print(df.info())

df.head()
print("\nDataset Description:")
print(df.describe())

# Step 2: Check for Duplicate Rows
duplicates = df[df.duplicated()]
print(f"\nDuplicate Rows: {len(duplicates)}")

# Step 3: Check for Duplicate Values in a Column (e.g., Patient_ID)
duplicate_values = df['Patient ID'].duplicated().sum()
print(f"\nDuplicate values in Patient ID: {duplicate_values}")

# Step 4: Calculate Mean, Median, and Mode for Numeric Columns
numeric_cols = df.select_dtypes(include=[np.number])

# Calculate mean, median, and mode
means = numeric_cols.mean()
medians = numeric_cols.median()
modes = numeric_cols.mode().iloc[0]

print("\nMeans:\n", means)
print("\nMedians:\n", medians)
print("\nModes:\n", modes)

# Step 5: Check for Missing or Null Values
missing_values = df.isnull().sum()
print(f"\nMissing values:\n {missing_values}")

# Handling missing values
df.dropna(inplace=True)

# Step 6: Identify Inconsistent Data
inconsistent_data = df[df['Age at Diagnosis'] < 0]  
print(f"\nInconsistent data:\n {inconsistent_data}")

# Step 7: Encode Categorical Variables
# One-hot encoding for categorical variables
df_encoded = pd.get_dummies(df, drop_first=True)  
print(f"\nEncoded Data (Head):\n {df_encoded.head()}")

```
